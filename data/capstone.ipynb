```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
```

# Data regarding males between 30-54 y/o

```python
ages_has_team = pd.read_excel('age_breakdown.xlsx',sheet_name='Has MLB Team')
```

```python
ages_no_team = pd.read_excel('age_breakdown.xlsx',sheet_name='NO MLB Team')
```

```python
ages_has_team
```

```python
ages_no_team
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
sns.barplot(x = "Percent of Male Pop Between 30-54 Years Old",
            y = "City",
           data = ages_has_team)
#ax.xaxis.set_major_formatter(mtick.PercentFormatter())
ax.set_xticklabels(map('{:.1f}%'.format, 100*ax.xaxis.get_majorticklocs()));
```

```python
ages_has_team.mean()
```

```python
cols = ['grey' if y < .179 else 'green' for y in ages_no_team["Percent of Male Pop Between 30-54 Years Old"]]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(x = "Percent of Male Pop Between 30-54 Years Old",
            y = "City", palette= cols,
           data = ages_no_team).axvline(x=0.1789, color = 'black', linewidth=2,)
ax.text(0.1835,4, "Avg MLB City 17.9%")
#.text(x= 0, y=9.25, s='target');
#plt.axvline(x=np.nanmean(ages_has_team.x))
ax.set_xticklabels(map('{:.1f}%'.format, 100*ax.xaxis.get_majorticklocs()));
```
